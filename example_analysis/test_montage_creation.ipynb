```python
import pandas as pd

from lib.aggregate.load_format_data import (
    load_hdf_subset,
    add_filenames,
)
```

```python
df_mitotic = load_hdf_subset(
    "/lab/barcheese01/rkern/brieflow/example_analysis/analysis_root/aggregate_process/hdfs/mitotic_data.hdf5",
    n_rows=5000,
)
df_mitotic
```

```python
from pathlib import Path

df_mitotic_filenames = add_filenames(df_mitotic, Path("/lab/barcheese01/rkern/brieflow/example_analysis/analysis_root"), montage_subset=True)
df_mitotic_filenames
```

```python
from lib.aggregate.montage_utils import create_cell_montage

montage_data = df_mitotic_filenames[(df_mitotic_filenames["gene_symbol_0"] == "ZNF728") & (df_mitotic_filenames["sgRNA_0"] == "AAGGACTGAGACCCAGCTAA")]
montage_data

# TODO: Remove when done with testing on Denali data
# redirects actual paths to paths of Denali data
def update_file_path(row):
    original_path = row["image_path"]
    parts = original_path.split("/")[-1].split("__")[0].split("_")
    well, tile = parts[0][1:], parts[1][1:]
    new_path = f"/archive/cheeseman/OpticalPooledScreens/lab/barcheese01/screens/denali/input_ph_tif/20X_{well}_Tile-{tile}.phenotype.tif"
    return new_path


montage_data["image_path"] = montage_data.apply(update_file_path, axis=1)

montage = create_cell_montage(
    montage_data,
    channels = ["DAPI", "COXIV", "CENPA", "WGA"],
)
test = montage["DAPI"]
test
```
